```python
from PIL import Image
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
root_data_folder = "8bit_characters"
image_names = [root_data_folder + "/" + name for name in os.listdir(root_data_folder) if name.endswith(".jpg")]
print(image_names)
```

```python
def plot_single_image(img):
    plt.imshow(img)
    plt.show()
```

```python
img = Image.open(image_names[0])
new_img = img.resize((250, 250)) # this is the half size of the image
width, height = new_img.size
padding = 3

result = Image.new(new_img.mode, (width + padding * 2, height + padding * 2), "red")
result.paste(new_img, (padding, padding))
plot_single_image(result)
print(result.size)
```

```python
def make_256_save(img_name):
    img = Image.open(img_name)
    new_img = img.resize((250, 250)) # this is the half size of the image
    width, height = new_img.size
    padding = 3

    result = Image.new(new_img.mode, (width + padding * 2, height + padding * 2), "white")
    result.paste(new_img, (padding, padding))
    if result.size != (256,256):
        raise Exception("Image size is not 256x256")

    real_name = img_name.split("/")[-1]
    
    result.save("flat_256/" + real_name, quality=100)
```

```python
for name in image_names:
    make_256_save(name)
```

```python
area = (50, 50, 500-50, 500-50)
cropped_img = img.crop(area)
```

```python
plot_single_image(cropped_img)
```
